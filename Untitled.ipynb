```python
import numpy as np 
import matplotlib.pyplot as plt
import oritelib as orite
```

```python
seq = orite.seq_from_fasta('test_data/pasteurella_multocida.fasta')
#
```

```python
# x = [final_gc, final_cgc, max_rotated_fasta, max_cCG_indx_original_offset]
x_25k = orite.max_rotate_seq_and_skew_calc(f=seq, window_radius=25000)
x_50k = orite.max_rotate_seq_and_skew_calc(f=seq, window_radius=50000)
x_100k = orite.max_rotate_seq_and_skew_calc(f=seq, window_radius=100000)
x_200k = orite.max_rotate_seq_and_skew_calc(f=seq, window_radius=200000)
```

```python
plt.figure(figsize=[40,10])
plt.plot(x_25k[0], 'k')
plt.plot(x_50k[0], 'b')
plt.plot(x_100k[0], 'g')
plt.plot(x_200k[0], 'r')

plt.axhline(y=0)
```

```python
plt.figure(figsize=[40,10])

plt.plot(x_25k[1], 'k', label='25k')
plt.plot(x_50k[1], 'b', label = '50k')
plt.plot(x_100k[1], 'g', label = '100k') 
plt.plot(x_200k[1], 'r', label = '200k')

plt.axhline(y=0)
plt.axhline(y=0)
plt.legend()
```
