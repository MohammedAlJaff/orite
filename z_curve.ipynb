```python
import oritelib as orite
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import minmax_scale
```

```python
# import seq 

seq = orite.seq_from_fasta('test_data/exciting_bacteria/Porphyromonas_gingivalis_L3_FASTA.fasta')
max_stuff = orite.max_rotate_seq_and_skew_calc(seq)
```

```python
cum_base_count = orite.cumilative_base_count(max_stuff[2])
```

```python
n = np.array(list(range(len(seq))))
```

```python
np.shape(cum_base_count)
```

```python
an = cum_base_count[0,:]
cn = cum_base_count[1,:]
gn = cum_base_count[2,:]
tn = cum_base_count[3,:]
```

```python
xn = (an+gn) - (cn+tn) #RY
yn = (an+cn) - (gn+tn) # MK
zn = (an+tn) - (cn+gn) 
```

```python
# rescale : 

x_scaled = minmax_scale(X=xn, feature_range=(-1,1))
y_scaled = minmax_scale(X=yn, feature_range=(-1,1))
z_scaled = minmax_scale(X=zn, feature_range=(-1,1))
max_cgc_scaled = minmax_scale(X=max_stuff[1], feature_range=(-1,1))




```

```python
plt.figure(figsize=[20,10])
plt.plot(n, x_scaled, label='x / RY')
plt.plot(n, (-1)*y_scaled, label='y / MK')
#plt.plot(n, z_scaled, label='z')
plt.plot(n, max_cgc_scaled, label='max_cgc')
plt.legend()


```

```python


q = orite.calc_z_curve(max_stuff[2])
```

```python
plt.figure(figsize=[20,10])
plt.plot(n, q[0], label='x / RY')
plt.plot(n, (-1)*q[1], label='y / MK')
plt.plot(n, max_stuff[1], label='z')
plt.plot(n, max_cgc_scaled, label='max_cgc')
plt.legend()
```
